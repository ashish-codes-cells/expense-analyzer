```python
import pandas as pd
```

```python
df = pd.read_csv("expenses.csv")
```

```python
df
```

```python
grouped_df = df.groupby("Category")["Amount"].agg(Amount="sum").reset_index()
```

```python
grouped_df
```

```python
print(grouped_df["Amount"].sum()) # Total expenses

grouped_df[grouped_df["Amount"] == grouped_df["Amount"].max()] # Highest expense

grouped_df[grouped_df["Amount"] == grouped_df["Amount"].min()] # Lowest expense
```

# Insights

1. The total expenses are rs.1410.
2. I'm spending the most on food and shoping (500 on food and 500 on shoping), which is the major part of total expenses.
3. My lowest expense is 160. I'm spending the least amount on book.
4. If the cost of shoping is reduced a little, then there can be more savings.

```python
import matplotlib.pyplot as plt

category = grouped_df["Category"]
amount = grouped_df["Amount"]

plt.bar(category, amount)
plt.title("Expense Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Amount")
```

The graph clearly indicates that spending on food and shoping is high,
while the lowest amount is being spent on books.
Expenditure on shoping can be reduced, as doing so would increase savings.
